## **SCRAPPING DATA**

```python
API_KEY = "---"

VIDEO_URLS = [
    "https://www.youtube.com/watch?v=J4UUQvi3FBY",
    "https://www.youtube.com/watch?v=RYwFpsaeHDQ",
    "https://www.youtube.com/watch?v=bfjxOOUyLR8",
    "https://www.youtube.com/watch?v=YSMuusyVR14",
]

OUTPUT_FILE = "youtube_ai_pendidikan_indonesia.xlsx"
```

```python
def extract_video_id(url):
    return url.split("v=")[-1]

def clean_text(text):
    text = re.sub(r"http\S+|www\S+", "", text)  # remove links
    text = re.sub(r"\s+", " ", text).strip()
    return text

def is_valid_comment(text):
    words = text.split()

    # Minimum word length
    if len(words) < 8:
        return False

    # Alphabet ratio (anti emoji spam)
    alpha_ratio = sum(c.isalpha() for c in text) / max(len(text), 1)
    if alpha_ratio < 0.6:
        return False

    # Repeated spam pattern
    if re.search(r"(.)\1{4,}", text):
        return False

    return True
```

```python
youtube = build(
    "youtube",
    "v3",
    developerKey=API_KEY
)

rows = []
```

```python
for url in VIDEO_URLS:
    video_id = extract_video_id(url)

    video_info = youtube.videos().list(
        part="snippet",
        id=video_id
    ).execute()

    snippet = video_info["items"][0]["snippet"]
    video_title = snippet["title"]
    published_at = snippet["publishedAt"]

    published_dt = datetime.fromisoformat(published_at.replace("Z", ""))
    publish_year = published_dt.year
    publish_month = published_dt.month

    next_page_token = None

    while True:
        response = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=100,
            pageToken=next_page_token,
            textFormat="plainText"
        ).execute()

        for item in response["items"]:
            top_comment = item["snippet"]["topLevelComment"]["snippet"]
            text = clean_text(top_comment["textDisplay"])

            if not is_valid_comment(text):
                continue

            rows.append({
                "author_id": top_comment.get("authorChannelId", {}).get("value"),
                "comment_id": item["snippet"]["topLevelComment"]["id"],
                "comment_text": text,
                "like_count": top_comment.get("likeCount", 0),
                "reply_count": item["snippet"].get("totalReplyCount", 0),
                "video_title": video_title,
                "publish_year": publish_year,
                "publish_month": publish_month
            })

        next_page_token = response.get("nextPageToken")
        if not next_page_token:
            break
```

```python
df = pd.DataFrame(rows)
df.to_excel(OUTPUT_FILE, index=False)

print(f"Saved {len(df)} comments to {OUTPUT_FILE}")
```

#### **CHECK NOISE**

```python
def check_text_noise(text: str) -> dict:
    if not isinstance(text, str):
        return {
            "has_html": False,
            "has_url": False,
            "has_hashtag": False,
            "has_mention": False,
            "has_number": False,
            "has_repeated_char": False,
            "has_non_ascii": False
        }

    return {
        "has_html": bool(re.search(r"<[^>]+>", text)),
        "has_url": bool(re.search(r"http[s]?://|www\.", text)),
        "has_hashtag": bool(re.search(r"#\w+", text)),
        "has_mention": bool(re.search(r"@\w+", text)),
        "has_number": bool(re.search(r"\d+", text)),
        "has_repeated_char": bool(re.search(r"(.)\1{2,}", text)),  # contoh: loooool
        "has_non_ascii": bool(re.search(r"[^\x00-\x7F]", text))
    }


def noise_eda_dataframe(df: pd.DataFrame, text_col: str) -> pd.DataFrame:
    noise_features = df[text_col].apply(check_text_noise)
    noise_df = pd.DataFrame(list(noise_features))
    summary = noise_df.mean().sort_values(ascending=False)
    return noise_df, summary
```

#### **CHECK TIMESTAMP FORMAT**

```python
TIMESTAMP_PATTERN = re.compile(
    r'\b(?:\d{1,2}:)?\d{1,2}:\d{2}\b'
)

def eda_check_timestamp(df: pd.DataFrame, text_col: str):
    temp = df[text_col].astype(str)

    result = pd.DataFrame({
        "has_timestamp": temp.apply(lambda x: bool(TIMESTAMP_PATTERN.search(x))),
        "timestamp_found": temp.apply(
            lambda x: TIMESTAMP_PATTERN.findall(x)
        )
    })

    summary = {
        "total_rows": len(df),
        "rows_with_timestamp": int(result["has_timestamp"].sum()),
        "percentage": round(result["has_timestamp"].mean() * 100, 2)
    }

    return result, summary
```

#### **CHECK KALIMAT NON-EYD & BAHASA ASING**

```python
kbbi_df = pd.read_csv("/kaggle/input/datasetnomor2/kbbi_data.csv", sep="\t")
kbbi_words = set(kbbi_df["katakunci"].str.lower())
```

```python
def eda_non_kbbi_tokens(text_series, kbbi_words, min_freq=3):
    counter = Counter()

    for text in text_series.dropna():
        tokens = re.findall(r"[a-zA-Z]+", text.lower())
        for tok in tokens:
            if tok not in kbbi_words:
                counter[tok] += 1

    non_kbbi_df = (
        Counter(counter)
        .most_common()
    )

    return [
        (tok, freq)
        for tok, freq in non_kbbi_df
        if freq >= min_freq
    ]
```

```python
non_kbbi_tokens = eda_non_kbbi_tokens(
    df["comment_text"],
    kbbi_words,
    min_freq=5
)
```

```python
def tokenize(text):
    return re.findall(r"[a-zA-Z]+", text.lower())

all_tokens = []
for t in df["comment_text"]:
    all_tokens.extend(tokenize(t))

token_freq = Counter(all_tokens)
```

```python
non_kbbi_tokens = {
    token: freq
    for token, freq in token_freq.items()
    if token not in kbbi_words
}
```

```python
def is_non_standard(token):
    if token in kbbi_words:
        return False
    if len(token) <= 3:      # singkatan informal
        return True
    if token.endswith(("nya", "in")):  # bentuk lisan
        return True
    return False
```

#### **CHECK EMOJI**

```python
EMOJI_PATTERN = re.compile(
    "["
    "\U0001F1E6-\U0001F1FF"  # flags
    "\U0001F600-\U0001F64F"
    "\U0001F300-\U0001F5FF"
    "\U0001F680-\U0001F6FF"
    "\U0001F700-\U0001F77F"
    "\U0001F780-\U0001F7FF"
    "\U0001F800-\U0001F8FF"
    "\U0001F900-\U0001F9FF"
    "\U0001FA00-\U0001FAFF"
    "\U00002702-\U000027B0"
    "\U000024C2-\U0001F251"
    "]",
    flags=re.UNICODE
)

def extract_emojis(text):
    return EMOJI_PATTERN.findall(text)

def collect_all_emojis(series):
    all_emojis = []
    for text in series:
        all_emojis.extend(extract_emojis(text))
    return all_emojis

def print_emoji_summary(series, top_n=20):
    all_emojis = collect_all_emojis(series)
    counter = Counter(all_emojis)

    print(f"Total emoji ditemukan: {sum(counter.values())}")
    print(f"Jumlah emoji unik: {len(counter)}")
    print("\nTop emoji:")
    for emoji, freq in counter.most_common(top_n):
        print(f"{emoji}  -> {freq}")

    return counter
```

#### **CHECK REPEATING SYMBOL**

```python
REPEATING_SYMBOL_REGEX = re.compile(r"([^\w\s])\1{1,}")

def eda_repeating_symbols(text):
    if not isinstance(text, str):
        return []
    return [m.group(0) for m in REPEATING_SYMBOL_REGEX.finditer(text)]
```

```python
def extract_numbers(text):
    if not isinstance(text, str):
        return []
    return re.findall(r"\b\d+\b", text)

def eda_numbers(series, top_n=20):
    all_numbers = []
    for text in series:
        all_numbers.extend(extract_numbers(text))

    counter = Counter(all_numbers)

    print(f"Total angka unik: {len(counter)}")
    print(f"Total kemunculan angka: {sum(counter.values())}")
    print("\nTop angka:")
    for num, freq in counter.most_common(top_n):
        print(f"{num} -> {freq}")

    return counter
```

#### **HANDLE SYMBOLS**

```python
def normalize_repeating_symbols(text):
    if not isinstance(text, str):
        return text
    
    # ... ,,, !!! ??? → satu simbol
    text = re.sub(r"([.!?,]){2,}", r"\1", text)
    
    # ----, —— → -
    text = re.sub(r"[-—]{2,}", "-", text)

    return text
```

```python
PURE_SYMBOL_CLEAN_PATTERN = re.compile(r"[^\w\s]", flags=re.UNICODE)

def remove_pure_symbols(text):
    if not isinstance(text, str):
        return text
    
    return PURE_SYMBOL_CLEAN_PATTERN.sub(" ", text)
```

```python
def clean_weird_symbols(text):
    if not isinstance(text, str):
        return text

    text = normalize_repeating_symbols(text)
    text = remove_pure_symbols(text)
    text = re.sub(r"\s+", " ", text).strip()
    
    return text
```

```python
df["comment_text"] = df["comment_text"].apply(clean_weird_symbols)
```

```python
BASE_STOPWORDS_ID = {
    "dan","yang","di","ke","dari","pada","dengan","untuk","oleh","dalam",
    "atau","karena","jadi","juga","hanya","akan","adalah","itu","ini",
    "ada","sudah","belum","tidak","bukan","lebih","kurang","semua",
    "apa","saja","menurut","kalau","agar","maka","harus"
}

NON_EYD_STOPWORDS = {
    "yg","jg","dgn","utk","aja","sdh","tdk","gak","ga","klo","krn",
    "tp","hrs","jd","org","sy","gw","lu","bs","gk","pd","dll",
    "dr","dlm","spt","ttg","udh","cm","nih","yah","deh"
}

PRONOUN_STOPWORDS = {
    "saya","kita","kami","mereka","dia","aku","anda","kalian",
    "pak","bu","mas","mbak"
}
FILLER_STOPWORDS = {
    "aja","nih","loh","lah","kok","sih","yah","nah","kan"
}

TAMBAHAN_SAPAAN = {
    "helmy", "yahya", "davyn", "sudirdjo",
    "elon", "musk", "gibran", "jansen",
    "jason", "wilson", "davin", "pak", "bu", "mas", 
    "mbak", "bang", "kak", "koh", "bung", "prof", "doktor", 
    "dr", "imo", "guys", "gan", "bro", "sis", "sob"
}

TFIDF_STOPWORDS = (
    BASE_STOPWORDS_ID
    | NON_EYD_STOPWORDS
    | PRONOUN_STOPWORDS
    | FILLER_STOPWORDS
    | TAMBAHAN_SAPAAN
)
```

```python
def remove_tfidf_stopwords(text, stopwords):
    if not isinstance(text, str):
        return ""
    tokens = text.split()
    tokens = [t for t in tokens if t not in stopwords]
    return " ".join(tokens)
```

```python
def top_terms_per_cluster_tfidf(
    vectorizer, X_tfidf, labels, top_n=15
):
    terms = vectorizer.get_feature_names_out()
    clusters = {}

    for k in sorted(set(labels)):
        idx = np.where(labels == k)[0]
        centroid = X_tfidf[idx].mean(axis=0)
        centroid = np.asarray(centroid).ravel()

        top_idx = centroid.argsort()[::-1][:top_n]
        clusters[k] = [terms[i] for i in top_idx]

    return clusters
```

```python
tfidf_cluster_terms = top_terms_per_cluster_tfidf(
    tfidf_vectorizer,
    X_tfidf,
    df["cluster_tfidf"].values,
    top_n=15
)

tfidf_cluster_terms
```

#### **KMEANS + IndoBERT**

```python
def representative_texts_indobert(
    embeddings, labels, texts, top_n=5
):
    result = {}

    for k in sorted(set(labels)):
        idx = np.where(labels == k)[0]
        cluster_emb = embeddings[idx]
        centroid = cluster_emb.mean(axis=0, keepdims=True)

        sims = cosine_similarity(cluster_emb, centroid).ravel()
        top_idx = idx[sims.argsort()[::-1][:top_n]]

        result[k] = [texts[i] for i in top_idx]

    return result
```
